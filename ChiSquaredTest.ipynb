```python
import numpy as np
import random
```

Chi-Squared Test

```python
lst = []
arr = []

row = 0
while row < 1000:
  for i in range(10):
    lst.append(np.random.normal(0,1))
  arr.append(lst)
  row += 1

#print(arr[0][1:10])

lst_row = []
for i in range(len(arr)):
  sq_sum = 0
  for j in range(len(arr[i])):
    sq_sum += (arr[i][j]**2)
  lst_row.append(sq_sum)

print(lst_row)
mean = sum(lst_row)/len(lst_row)
variance = sum((i - mean) ** 2 for i in lst_row) / len(lst_row)

print(mean,variance, sep = ' ')
```
